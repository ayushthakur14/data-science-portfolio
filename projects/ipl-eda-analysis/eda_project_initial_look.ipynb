## Dataset Overview

- Ball-by-ball IPL data, 2008–2025 — 283,678 rows x 65 columns
- Granularity: one row per ball bowled
- No duplicate rows
- `next_batter` is mostly null by design (only filled when a wicket falls) — not a data quality issue
- `Unnamed: 0` is a junk index column, to be dropped later
- Other missing values likely concentrated in wicket-related and DRS-related columns, since those only apply when relevant events happen

## Questions to Explore

1. Does winning the toss actually correlate with winning the match, and does that vary by venue?
2. Which teams/players perform best in run-chases vs defending totals?
3. How has scoring rate (runs per over) evolved across seasons — has the game gotten more aggressive over time?

```python
# Q1 Does winning the toss actually correlate with winning the match, and does that vary by venue?
df['toss_winner_won'] = df['toss_winner'] == df['match_won_by']
df['toss_winner_won'].value_counts(normalize=True)
```

Winning the toss provided almost no advantage, toss winners won 50.5% of matches, essentially identical to random chance (50%). This suggests the toss itself is not a meaningful predictor of match outcome across the full dataset.

```python

df.groupby('venue')['match_id'].nunique().sort_values(ascending=False).head(10)
```

A few venues appeared to show 100% toss-win correlation, but this is a small-sample artifact, venues with only 1-2 matches show extreme percentages by chance. Venues with substantial match history stay close to the overall 50.5% baseline, confirming that toss provides no meaningful advantage anywhere in this dataset.

```python
import matplotlib.pyplot as plt
```

```python
df.drop_duplicates('match_id')['toss_decision'].value_counts().plot(kind='bar')
plt.title('Toss Decision: Bat vs Field (per match)')
plt.xlabel('Decision')
plt.ylabel('Count')
plt.show()
```

After correcting for ball-level duplication (using one row per match), captains chose to field first in the significant majority of matches (~780 vs ~400), reflecting the common T20 strategy of chasing a target rather than defending one.

```python
df_matches = df.drop_duplicates('match_id')
df_matches.groupby('toss_decision')['toss_winner_won'].mean()
```

While winning the toss overall showed no advantage, the decision made after winning matters more: toss winners who chose to field first won 53.7% of matches, versus 45.2% when choosing to bat first. This supports the common IPL strategy of chasing, and suggests captains' preference for fielding first (~780 vs ~400 matches) is a statistically reasonable strategy, not just a fashion trend.

```python
df.drop_duplicates('match_id')['match_won_by'].value_counts().head(10).plot(kind='bar', figsize=(10,5))
plt.title('Top 10 Teams by Matches Won')
plt.xlabel('Team')
plt.ylabel('Wins')
plt.xticks(rotation=45)
plt.show()
```

Mumbai Indians, Chennai Super Kings, and Kolkata Knight Riders lead in total wins, but this is partly a function of longevity, not just performance. Newer franchises (Gujarat Titans, Lucknow Super Giants if present) have had far fewer seasons to accumulate wins, so total win count alone is not a fair performance comparison across teams. A win-percentage or wins-per-season metric would be more accurate.

```python
df.groupby('season')['runs_total'].sum().plot(kind='line' , figsize = (10,5))
plt.title('Total Runs per Season')
plt.xlabel('Season')
plt.ylabel('Total Run')
plt.xticks(rotation=45)
plt.show()
```

Total runs per season show a general upward trend from 2008 to 2025, suggesting the game has become more aggressive/higher-scoring over time. The sharp drop in 2026 is not a real decline in scoring - verified that the 2026 season only has 24 matches recorded in this dataset, compared to 74 in the full 2025 season. This confirms a data completeness gap for 2026, not an actual cricketing trend.

```python
# Q2 Which teams/players perform best in run-chases vs defending totals?
batting_first = df[df['innings'] == 1].drop_duplicates('match_id')[['match_id', 'batting_team', 'match_won_by']]

batting_first['defended_successfully'] = batting_first['batting_team'] == batting_first['match_won_by']

batting_first['defended_successfully'].value_counts(normalize=True)
```

Teams batting first (defending) won only 44.8% of matches, while teams batting second (chasing) won 55.2%. This confirms the common cricketing wisdom that chasing is generally advantageous in T20 cricket likely due to having a clear target to chase rather than needing to set one, and conditions (like dew) often favoring the team bowling first in the second innings.

```python
batting_first.groupby('batting_team')['defended_successfully'].mean().sort_values(ascending=False)
```

```python
batting_first['batting_team'].value_counts()
```

Among established teams with substantial match history, Mumbai Indians and Chennai Super Kings show strong, consistent defending records (~52%). Teams like Rising Pune Supergiant(s) show extreme percentages, but this is unreliable due to very few matches and inconsistent team-name spelling in the raw data (a data quality issue, not a real performance signal)

```python
# Q3 How has scoring rate (runs per over) evolved across seasons — has the game gotten more aggressive over time? 

season_runs = df.groupby('season').agg(
    total_runs=('runs_total', 'sum'),
    total_balls=('ball_no', 'count')
)
season_runs['runs_per_over'] = (season_runs['total_runs'] / season_runs['total_balls']) * 6
season_runs['runs_per_over']
```

```python
season_runs['runs_per_over'].plot(kind='line', figsize=(10,5), marker='o')
plt.title('Average Runs per Over by Season')
plt.xlabel('Season')
plt.ylabel('Runs per Over')
plt.xticks(rotation=45)
plt.show()
```

Using runs-per-over (rather than total runs) confirms the game has become genuinely more aggressive over time, average scoring rose from ~7.98 runs/over in 2007/08 to ~9.39 in 2026. Unlike the total-runs chart, this rate-based view shows no artificial 2026 dip, confirming the earlier finding that the drop in total runs was a data completeness issue, not a real decline. Notably, the sharpest rise happened recently — from ~7.7 runs/over in 2021 to over 9.0 by 2024 onward, suggesting a real shift in playing style in the last few seasons (possibly linked to rule changes, bigger bats, or shorter boundaries at newer venues), rather than a gradual 18-year trend.
