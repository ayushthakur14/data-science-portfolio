```python
# Employee Performance

# Group by Department and calculate the average Salary and maximum Experience using .agg().
# Create a new column named Department Average Salary using .transform().
```

```python
import pandas as pd

df = pd.DataFrame({
    "Department": ["IT", "IT", "HR", "HR", "Sales", "Sales"],
    "Team": ["A", "A", "B", "B", "C", "C"],
    "Salary": [60000, 70000, 50000, 45000, 55000, 65000],
    "Experience": [3, 5, 6, 2, 4, 7]
})

print(df)
```

```python
df.groupby("Department").agg({"Salary" : "mean" , "Experience" : "max"})
```

```python
df["Department Average Salary"] = df.groupby("Department")["Salary"].transform("mean")
print(df)
```
