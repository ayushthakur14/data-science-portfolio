```python
# Convert the array to int32 and determine how many bytes of memory are saved after the conversion.

import numpy as np

arr = np.arange(2000000, dtype=np.int64)

print("Original Memory:", arr.nbytes, "bytes")

arr32 = arr.astype(np.int32)

print("New Memory:", arr32.nbytes, "bytes")

print("Memory Saved:", arr.nbytes - arr32.nbytes, "bytes")
```
