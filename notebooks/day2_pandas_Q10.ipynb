```python
# Company Salaries

# Find the average salary of each department.
# Find the maximum salary of each team.
```

```python
import pandas as pd

df = pd.DataFrame({
    "Department": ["IT", "IT", "HR", "HR", "Sales", "Sales"],
    "Team": ["A", "A", "B", "B", "C", "C"],
    "Salary": [60000, 70000, 50000, 45000, 55000, 65000]
})

print(df)
```

```python
df.groupby("Department")["Salary"].mean()
```

```python
df.groupby("Team").max()
```
