```python
# Compare the time taken to multiply two Python lists vs. two NumPy arrays 

import numpy as np
import time

size = 1_000_000

l1 = list(range(size))
l2 = list(range(size))

start = time.time()
muliply = [x*y for x,y in zip(l1, l2)]
end = time.time()

print(multiply[0:10])
print(end - start)
```

```python
size = 1_000_000

arr1 = np.arange(size)
arr2 = np.arange(size)

start = time.time()
multiply = arr1 * arr2 
end = time.time()

print(multiply[0:10])
print(end - start)
```
