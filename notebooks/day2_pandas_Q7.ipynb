```python
import pandas as pd

df = pd.DataFrame({
    "Name": [" Alice ", "Bob ", " Charlie", "DAVID", "Eva"],
    "City": ["Delhi", "mumbai", "DELHI", "Pune", "bangalore"]
})

print(df)
```

```python
df["Name"] =  df["Name"].str.strip()
print(df)
```

```python
df["City"] =  df["City"].str.upper()
print(df)
```
