```python
# Employee & Department Data

# Perform a left merge using DeptID.
# Display employees whose department information is missing after the merge.
```

```python
import pandas as pd

employees = pd.DataFrame({
    "EmpID": [1, 2, 3, 4],
    "Name": ["Alice", "Bob", "Charlie", "David"],
    "DeptID": [101, 102, 103, 104]
})

departments = pd.DataFrame({
    "DeptID": [101, 102, 104],
    "Department": ["HR", "IT", "Marketing"]
})

```

```python
merged = pd.merge(employees , departments , on = "DeptID" , how = "left")
print(merged)
```

```python
missing_dept = merged[merged["Department"].isna()]
print(missing_dept)
```
