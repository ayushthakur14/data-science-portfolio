```python
# Student Marks

# Convert the DataFrame into long format using melt().
# Rename the generated columns as Subject and Marks.
```

```python
import pandas as pd

df = pd.DataFrame({
    "Name": ["Alice", "Bob", "Charlie"],
    "Math": [90, 75, 95],
    "Science": [88, 82, 94],
    "English": [91, 79, 96]
})

print(df)
```

```python
long_df = df.melt(
    id_vars="Name",
    var_name="Subject",
    value_name="Marks"
)

print(long_df)
```
