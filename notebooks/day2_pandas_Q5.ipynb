```python
# Missing Data

# Count the missing values in each column.
# Replace missing values with the mean of their respective columns.
```

```python
import pandas as pd
import numpy as np

df = pd.DataFrame({
    "Name": ["A", "B", "C", "D", "E"],
    "Marks": [85, np.nan, 90, np.nan, 75],
    "Attendance": [95, 80, np.nan, 90, 88]
})

print(df)
```

```python
df.isnull().sum()
```

```python
df = df.fillna(df.mean(numeric_only = True))
print(df)
```
