```python
# Student Categories

# Create a new column named Category (Adult if Age ≥ 18, otherwise Minor).
# Replace M with Male and F with Female.
```

```python
import pandas as pd

df = pd.DataFrame({
    "Age": [15, 18, 22, 16, 25],
    "Gender": ["M", "F", "M", "F", "M"]
})

print(df)
```

```python
df["Category"] = df["Age"].apply(lambda x : "Adult" if x >= 18 else "Minor")
print(df)
```

```python
df["Gender"] = df["Gender"].replace({"M" : "Male" , "F" : "Female"})
print(df)
```
