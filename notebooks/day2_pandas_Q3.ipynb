```python
# Employee Records

# Display employees whose salary is greater than 50,000.
# Display only employees from the IT department.
```

```python
import pandas as pd

df = pd.DataFrame({
    "Employee": ["John", "Sarah", "Mike", "Emma", "Alex"],
    "Salary": [40000, 65000, 55000, 72000, 38000],
    "Department": ["HR", "IT", "Sales", "IT", "HR"]
})

print(df)
```

```python
df[df["Salary"] > 50000]
```

```python
df[df["Department"] == "IT"]
```
