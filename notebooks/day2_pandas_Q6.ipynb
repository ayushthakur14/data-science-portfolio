```python
# Customer Database

# Remove duplicate rows.
# Sort the DataFrame by Purchase in descending order.
```

```python
import pandas as pd

df = pd.DataFrame({
    "Name": ["Alice", "Bob", "Alice", "David", "Bob", "Eva"],
    "City": ["Delhi", "Mumbai", "Delhi", "Pune", "Mumbai", "Chennai"],
    "Purchase": [500, 800, 500, 1200, 800, 650]
})

print(df)
```

```python
df.drop_duplicates()
```

```python
df.sort_values(by = "Purchase" , ascending = True)
```
